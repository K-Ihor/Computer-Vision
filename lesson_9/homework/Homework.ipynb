### Step 0

Run the necessary imports.


```python
# Import libraries
import dlib
from matplotlib import pyplot as plt
import cv2  # OpenCV for image reading and display
plt.rcParams['figure.figsize'] = [15, 10]
```

### Step 1

Load an image (any image that contains faces).

```python
# Step 1: Load an image 
image = cv2.imread("../data/eu_summit.jpg")
img = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

### Step 2

Load the dlib face predictor.

```python
# Load the dlib face detector model
detector = dlib.get_frontal_face_detector()
```

### Step 3

Run the predictor on your image.

```python
# Detect faces in the image
detected_faces = detector(image, 1)  # Upsample the image 1 time for better accuracy
```

### Step 4

Draw bounding boxes around the detected faces and plot the image. Use different colour for each face.

```python
# Draw bounding boxes and assign different colors for each face
color_list = [(255, 0, 0), (0, 255, 0), (0, 0, 255), (255, 255, 0)]  # List of colors (BGR)
count = 0
```

```python
for face in detected_faces:
    left = face.left()
    top = face.top()
    right = face.right()
    bottom = face.bottom()

    # Draw rectangle with different color for each face
    color = color_list[count % len(color_list)]  # Cycle through the color list
    cv2.rectangle(image, (left, top), (right, bottom), color, 2)
    count += 1  # Keep track of the number of faces

# Display the image with bounding boxes
cv2.imshow("Image with Face Detection", image)
cv2.waitKey(0)  # press Enter to close the window
cv2.destroyAllWindows()
```
